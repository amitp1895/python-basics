# Working with PANDAS

### set index

```python
df2=pd.DataFrame({"Days":[1,2,3,4], "visitor":[250,10,180,320], "bounce_rate":[50,41,67,88]})
print(df2)

#the change in the original data should be stored in another variable 
dfN=df2.set_index("Days")
dfN
```

```python
#if the changes are to be done in original data  the inplace can be used 
df2.set_index("Days", inplace=True)
print(df2)
```

Change the header 

```python
df2=df2.rename (columns={"visitor":"user"})
print(df2)
```
