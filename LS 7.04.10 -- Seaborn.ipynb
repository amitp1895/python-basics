### Category plot

Category plot takes one string and one numerical  input

```python
df=pd.read_csv("iris.csv")
print(df.head())
sns.set(style="whitegrid")
ax=sns.stripplot(x="class",y="sepal length",data=df)
plt.title( "Iris  strip plots ")
plt.show()
```

```python
ay=sns.swarmplot(x="class",y="sepal length",data=df)
plt.title("iris swarm plot ")
plt.show()
```

```python
#above plot in horizontal view 
az=sns.swarmplot(y="class",x=("sepal length"),data=df)
plt.title("iris swarmplot horizontal")
plt.show()
```
